OverSamppling :SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# Assuming 'data' is your DataFrame with 'Attr1' to 'Attr64' and 'Target' column
X = data.drop('class', axis=1)
y = data['class']

# Initialize the SMOTE object
smote = SMOTE()

# Perform SMOTE oversampling
X_resampled, y_resampled = smote.fit_resample(X, y)

# Combine the resampled features and target back into a DataFrame
resampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='class')], axis=1)
```

```python

from sklearn.impute import SimpleImputer
# Assuming 'data' is your DataFrame with 'Attr1' to 'Attr64' and 'Target' column
X = data.drop('class', axis=1)
y = data['class']

# Step 1: Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # You can choose different imputation strategies based on your data
X_imputed = imputer.fit_transform(X)

# Step 2: Apply SMOTE to oversample the target variable
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_imputed, y)

# Combine the resampled features and target back into a DataFrame
resampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='class')], axis=1)
```
